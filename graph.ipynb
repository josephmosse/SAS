```python
cmap = LinearSegmentedColormap.from_list("custom_cmap", ['#DBE2EF', blue])
cmap
```

```python
def geo_heatmap(gdf, var, cb=True):
        fig, ax = plt.subplots(figsize=(10, 10))
        gdf.plot(var,
                ax=ax,
                cmap=cmap,
                legend=cb,
                legend_kwds={'shrink':0.7, "fmt": "{:.0f}", 'location': "right", 'label':"Nombre de sociétaires"})
        ax.set_title('Départements de résidence des sociétaires')
        plt.axis('off')
        #plt.savefig('Diapo/img/'+var+'.png', bbox_inches='tight', pad_inches=0)
        plt.show()

geo_heatmap(gdf, 'Population')
```
